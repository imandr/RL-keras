```python
import numpy as np, generate
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sample(nx, nt, model):
    x = np.arange(-1.0, 1.0, 2.0/nx)
    t = np.arange(0.0, 1.0, 1.0/nt)
    xg, tg = np.meshgrid(x, t)
    s = np.array((xg.reshape((-1,)),tg.reshape((-1,)))).T
    print "s:",s.shape,s
    z = model.predict_on_batch(s)
    return xg, tg, z[:,0].reshape(xg.shape)


def plot(model):
    x,y,z = sample(30,30,model)
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    #surf = ax.plot_wireframe(x, y, z, linewidth=0, antialiased=False)
    surf = ax.plot_wireframe(x, y, z)
```
